```python
# Importing the required libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\train.csv")
```

## Numeric Features

```python
# Visualizing numeric features using sns regression plots.
numeric_features_set1 = ["MasVnrArea", "1stFlrSF", "GrLivArea", "GarageArea"]
for var in numeric_features_set1:
    sns.regplot(x=var, y="SalePrice", data=df)
    plt.title(f"{var} vs SalePrice")
    plt.xlabel(var)
    plt.ylabel("Sales Price")
    plt.show()
```

```python
# Visualizing numeric features using barplots.
numeric_features_set2 = ["BsmtFullBath", "BsmtHalfBath", "BedroomAbvGr", "KitchenAbvGr"]
for var in numeric_features_set2:
    sns.barplot(x=var, y="SalePrice", data=df, color="g", alpha=0.5)
    plt.xlabel(var)
    plt.ylabel("Sales Price")
    plt.show()
```

```python
# Visualizing numeric features using histplots.
numeric_features_set3 = ["LotFrontage", "LotArea", "BsmtUnfSF"]
for i in numeric_features_set3:
    data = df.copy()
    data[i].hist(bins=25)
    plt.xlabel(i)
    plt.ylabel("count")
    plt.show()
```

```python
# GrLivArea vs SalePrice [corr = 0.71]
Pearson_GrLiv = 0.71
plt.figure(figsize = (12,6))
sns.regplot(data=df, x = 'GrLivArea', y='SalePrice', scatter_kws={'alpha':0.2}, color="r")
plt.title('GrLivArea vs SalePrice', fontsize = 12)
plt.legend(['$Pearson=$ {:.2f}'.format(Pearson_GrLiv)], loc = 'best')
plt.show()
```

```python
# TotalBsmtSF vs SalePrice [corr = 0.63]
Pearson_TBSF = 0.63
plt.figure(figsize = (12,6))
sns.regplot(data=df, x = 'TotalBsmtSF', y='SalePrice', scatter_kws={'alpha':0.2}, color="y")
plt.title('TotalBsmtSF vs SalePrice', fontsize = 12)
plt.legend(['$Pearson=$ {:.2f}'.format(Pearson_TBSF)], loc = 'best')
plt.show()
```

```python
# YearBuilt vs SalePrice [corr = 0.56]
Pearson_YrBlt = 0.56
plt.figure(figsize = (12,6))
sns.regplot(data=df, x = 'YearBuilt', y='SalePrice', scatter_kws={'alpha':0.2}, color="b")
plt.title('YearBuilt vs SalePrice', fontsize = 12)
plt.legend(['$Pearson=$ {:.2f}'.format(Pearson_YrBlt)], loc = 'best')
plt.show()
```

## Categorical Features

```python
# OverallQuall - SalePrice [Pearson = 0.8]
figure, ax = plt.subplots(1,3, figsize = (20,8))
sns.stripplot(data=df, x = 'OverallQual', y='SalePrice', ax = ax[0])
sns.boxplot(data=df, x = 'OverallQual', y='SalePrice', ax = ax[1])
sns.violinplot(data=df, x = 'OverallQual', y='SalePrice', ax = ax[2])
plt.show()
```

```python
# TotRmsAbvGrd - SalePrice [Pearson = 0.50]
figure, ax = plt.subplots(1,3, figsize = (20,8))
sns.stripplot(data=df, x = 'TotRmsAbvGrd', y='SalePrice', ax = ax[0])
sns.boxplot(data=df, x = 'TotRmsAbvGrd', y='SalePrice', ax = ax[1])
sns.violinplot(data=df, x = 'TotRmsAbvGrd', y='SalePrice', ax = ax[2])
plt.show()
```

## Median of Sale Price by Year

```python
plt.figure(figsize = (10,5))
sns.barplot(x='YrSold', y="SalePrice", data = df, estimator = np.median)
plt.title('Median of Sale Price by Year', fontsize = 13)
plt.xlabel('Selling Year', fontsize = 12)
plt.ylabel('Median of Price in $', fontsize = 12)
plt.show()
```
